# Web Scraping Flipkart (Motivational Books Website)


```python
#importing reqd lib
from bs4 import BeautifulSoup
import requests as req
import pandas as pd
import numpy as np
```

```python
#URL for Flipkart website 
URL  = "https://www.flipkart.com/search?q=motivational%20books&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
#Setting user agent 
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36','Accept-Language':'en-US, en;q=0.5'})
```

```python
#sending request to the flipkart website
response = req.get(URL,headers = HEADERS)
```

```python
response #200 code means connection is OK
```

```python
#parsing the response object into HTML object
soup = BeautifulSoup(response.content,'html.parser')
```

```python
#defininf the functions to extract the specific details

def get_title(new_soup):
    
    try:
        
        title  = new_soup.find("span",attrs={"class":'B_NuCI'}).text.replace("\xa0\xa0"," ")
    except AttributeError:
        title = ""
    
    return title


def get_price(new_soup):
    
    try:
        
        price = new_soup.find("div",attrs={"class":'_30jeq3 _16Jk6d'}).text
    except AttributeError:
        price = ""
    
    return price


def get_rating(new_soup):
    
    try:
        
        rating = new_soup.find("div",attrs={"class":'_3LWZlK'}).text
    except AttributeError:
        rating = ""
    
    return rating

      
```

```python
#finding all the links with <a> tag
links = soup.find_all("a",{'class':'s1Q9rs'})

#creating empty list
links_list = []

for link in links:
    links_list.append(link.get('href'))

d = {"title":[],"price":[],"rating":[]}

#extracting web content of each product
for l in links_list:
    
    new_website = req.get("https://flipkart.com"+l,headers=HEADERS)
    new_soup = BeautifulSoup(new_website.content,'html.parser')
    
    #extracting each product info (like title,price and rating from the new_soup variable)
    d["title"].append(get_title(new_soup))
    d["price"].append(get_price(new_soup))
    d["rating"].append(get_rating(new_soup))


#converting the dictionary into the dataframe
flipkart_df = pd.DataFrame.from_dict(d)

    
```
